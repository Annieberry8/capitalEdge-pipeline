```python
import pandas as pd
import requests
import apache_beam as beam
from dotenv import load_dotenv
import os
import psycopg2
from sqlalchemy import create_engine
```

## Extraction of data

```python

load_dotenv()
api_key = os.getenv("Alpha_vantage_API_KEY")
api_key
```

```python
# URL Endpoint
symbol = "MSFT"
url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={symbol}&interval=5min&apikey={api_key}&outputsize=5"
```

```python
# make a request
response = requests.get(url)
response
```

```python
# get data in json
data = response.json()
data
```

```python
values = data['Time Series (Daily)']
values
```

```python
# Transform the data to a dataframe
def transform_data(data):
    # Extract values from json
    values = data['Time Series (Daily)']
    # convert to dataframe
    df = pd.DataFrame.from_dict(values, orient='index')
    # Convert the columns to appropriate data types
    # convert the index (date strings) to date
    df.index = pd.to_datetime(df.index)
    # Rename columns for clarity
    df.rename(columns={
    '1. open': 'open',
    '2. high': 'high',
    '3. low': 'low',
    '4. close': 'close',
    '5. volume': 'volume'
    }, inplace=True)
    
    df = df.astype({
        'open': 'float',
        'high': 'float',
        'low': 'float',
        'close': 'float',
        'volume': 'int'
    })
    # reset the index to make 'date' a column
    df.reset_index(inplace=True)
    df.rename(columns={'index': 'Date'}, inplace=True)
    df['Date'] = df['Date'].dt.date
    
    return df
```

```python
df = transform_data(data)
df
```

```python
df.to_csv("stock_prices_data.csv", index=False)
```

```python
# create pipeline with beam
input_file = "stock_prices_data.csv"
output_path = "output/processed"

with beam.Pipeline() as p:
    cleaned_stock_prices_data = (
        p
        | "Read Input File" >> beam.io.ReadFromText(input_file, skip_header_lines=1) 
    )   
```
